```python
# Import label encoder
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
df['engine_size']= label_encoder.fit_transform(df['engine_size'])
df['engine_size'].unique()

pd.to_numeric(df.loc[:,'engine_size'])

label_encoder = preprocessing.LabelEncoder()
df['make']= label_encoder.fit_transform(df['make'])
df['make'].unique()

pd.to_numeric(df.loc[:,'make'])

label_encoder = preprocessing.LabelEncoder()
df['engine_type']= label_encoder.fit_transform(df['engine_type'])
df['engine_type'].unique()

pd.to_numeric(df.loc[:,'engine_type'])

label_encoder = preprocessing.LabelEncoder()
df['fuel_type']= label_encoder.fit_transform(df['fuel_type'])
df['fuel_type'].unique()

pd.to_numeric(df.loc[:,'fuel_type'])

label_encoder = preprocessing.LabelEncoder()
df['aspiration']= label_encoder.fit_transform(df['aspiration'])
df['aspiration'].unique()

pd.to_numeric(df.loc[:,'aspiration'])


label_encoder = preprocessing.LabelEncoder()
df['num_of_doors']= label_encoder.fit_transform(df['num_of_doors'])
df['num_of_doors'].unique()

pd.to_numeric(df.loc[:,'num_of_doors'])


label_encoder = preprocessing.LabelEncoder()
df['body_style']= label_encoder.fit_transform(df['body_style'])
df['body_style'].unique()

pd.to_numeric(df.loc[:,'body_style'])


label_encoder = preprocessing.LabelEncoder()
df['drive_wheels']= label_encoder.fit_transform(df['drive_wheels'])
df['drive_wheels'].unique()

pd.to_numeric(df.loc[:,'drive_wheels'])


label_encoder = preprocessing.LabelEncoder()
df['engine_location']= label_encoder.fit_transform(df['engine_location'])
df['engine_location'].unique()

pd.to_numeric(df.loc[:,'engine_location'])


label_encoder = preprocessing.LabelEncoder()
df['wheel_base']= label_encoder.fit_transform(df['wheel_base'])
df['wheel_base'].unique()

pd.to_numeric(df.loc[:,'wheel_base'])

label_encoder = preprocessing.LabelEncoder()
df['length']= label_encoder.fit_transform(df['length'])
df['length'].unique()

pd.to_numeric(df.loc[:,'length'])

label_encoder = preprocessing.LabelEncoder()
df['width']= label_encoder.fit_transform(df['width'])
df['width'].unique()

pd.to_numeric(df.loc[:,'width'])

label_encoder = preprocessing.LabelEncoder()
df['height']= label_encoder.fit_transform(df['height'])
df['height'].unique()

pd.to_numeric(df.loc[:,'height'])

label_encoder = preprocessing.LabelEncoder()
df['curb_weight']= label_encoder.fit_transform(df['curb_weight'])
df['curb_weight'].unique()

pd.to_numeric(df.loc[:,'curb_weight'])

label_encoder = preprocessing.LabelEncoder()
df['engine_type']= label_encoder.fit_transform(df['engine_type'])
df['engine_type'].unique()

pd.to_numeric(df.loc[:,'engine_type'])

label_encoder = preprocessing.LabelEncoder()
df['num_of_cylinders']= label_encoder.fit_transform(df['num_of_cylinders'])
df['num_of_cylinders'].unique()

pd.to_numeric(df.loc[:,'num_of_cylinders'])

label_encoder = preprocessing.LabelEncoder()
df['engine_size']= label_encoder.fit_transform(df['engine_size'])
df['engine_size'].unique()

pd.to_numeric(df.loc[:,'engine_size'])

label_encoder = preprocessing.LabelEncoder()
df['fuel_system']= label_encoder.fit_transform(df['fuel_system'])
df['fuel_system'].unique()

pd.to_numeric(df.loc[:,'fuel_system'])

label_encoder = preprocessing.LabelEncoder()
df['bore']= label_encoder.fit_transform(df['bore'])
df['bore'].unique()

pd.to_numeric(df.loc[:,'bore'])

label_encoder = preprocessing.LabelEncoder()
df['stroke']= label_encoder.fit_transform(df['stroke'])
df['stroke'].unique()

pd.to_numeric(df.loc[:,'stroke'])

label_encoder = preprocessing.LabelEncoder()
df['compression_ratio']= label_encoder.fit_transform(df['compression_ratio'])
df['compression_ratio'].unique()

pd.to_numeric(df.loc[:,'compression_ratio'])

label_encoder = preprocessing.LabelEncoder()
df['horsepower']= label_encoder.fit_transform(df['horsepower'])
df['horsepower'].unique()

pd.to_numeric(df.loc[:,'horsepower'])

label_encoder = preprocessing.LabelEncoder()
df['peak_rpm']= label_encoder.fit_transform(df['peak_rpm'])
df['peak_rpm'].unique()

pd.to_numeric(df.loc[:,'peak_rpm'])

label_encoder = preprocessing.LabelEncoder()
df['city_mpg']= label_encoder.fit_transform(df['city_mpg'])
df['city_mpg'].unique()

pd.to_numeric(df.loc[:,'city_mpg'])

label_encoder = preprocessing.LabelEncoder()
df['highway_mpg']= label_encoder.fit_transform(df['highway_mpg'])
df['highway_mpg'].unique()

pd.to_numeric(df.loc[:,'highway_mpg'])

label_encoder = preprocessing.LabelEncoder()
df['prices']= label_encoder.fit_transform(df['prices'])
df['prices'].unique()

pd.to_numeric(df.loc[:,'prices'])
```

<p>The places we would use feature scaling method is in a few places like symbol, normalized losses, num of doors, bore, stroke, horsepower, peak rpm, and prices. Feature scaling can make our data more whole my filling in the empty gaps in our case the most affect is normalized loans because it has 40 missing numbers this can tilt our data signifinaclly the others not so much where num of doors horsepower and peak rpm only has 2 missing data points and prices bore, stroke has 4 missing where if we use our scaling it wont had any real value to our data.But the data that will be greatly affect is symbol due to its 60 missing values. <p>


<h1>#2</h1>
<p>There is 2 missing data points for num of doors and 4 missing with prices</p>
<p>The method we would use to fill the missing values is imputation where we exchange to empty values with a question mark, we will use. loc to retrieve our value and make it equal to the question mark and then use. count to add it in</p>


```python
#2
#we use imputation to exchange values of num of doors
df['num_of_doors'].loc[df['num_of_doors'] == '?'].count()
#we use imputation to exchange values for prices
df['prices'].loc[df['prices'] == '?'].count()
```

```python
#we will also use fillna prev values to fill the nans
df_fill1 = df.copy()
df_fill1.fillna(method='bfill', inplace=True)
print(df_fill1)
```
